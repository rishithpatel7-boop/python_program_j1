# operators in python

## python have 6 types of operators
1.arthematic operators
2.assigment operators
3.comparision operators
4.logical operators
5.membership operators
6.identity operators

```python
#arthematic operators= +,-,*,/,//,%
a=10
b=20
c=30
d=40
print(a+b)
print(b-d)
print(a*c)
print(b/a)
print(b//a)
print(d%b)
```

```python
#comporision opertors= >,<,>=,<=,==,!= (return true or false)
a=5
b=10
c=10
d=20
print(a>b)
print(b<c)
print(b>=c)
print(b==c)
print(b!=d)
```

```python
# logical operator = and, or, not
# and - this operators checking all conditions are true result is true rest cases failed
#or - this operator check either one conditions true  true then result true rest cases failed
a=10
b=20
c=20
d=60
print(a>b and b==c)
print(a<b and c<d)
print(a>b or b==c)
print(a>b or b!=c)
print(not a>b or b!=c)
```

```python
# membership opertor = checking members are present in that sequence or collection - in, not in
names=["shiva","vikas","divya","alok"]
print("alok" in names)
print("chandra" in names)
print("shiva" not in names)
print("vikas" in names)
```

```python
#identity operators -memory identification of values -is,is not
a=1000
b=1000
print(id(a))
print(id(b))
print(a is b)
```
